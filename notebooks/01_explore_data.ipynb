## Notes for Building a Mixed EN/FR Dataset

- SST-2 (EN) and Allociné (FR) both contain movie reviews, making them comparable in terms of task and content.
- Both datasets use compatible labels: 0 = negative, 1 = positive.
- Allociné is significantly larger than SST-2, so mixing them without control would heavily favor French samples.
- English texts are generally shorter, while French reviews tend to be longer and sometimes span multiple sentences.
- For fine-tuning the EN+FR teacher model, it will likely be necessary to either:
  - subsample the Allociné dataset, or
  - define a controlled mixing ratio (e.g., 30% EN and 70% FR).
- This notebook confirms that the two datasets are compatible for a bilingual sentiment classification task.
